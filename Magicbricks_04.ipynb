```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('MagicBricks.csv')
```

```python
df
```

```python
df.drop('Locality',axis=1,inplace=True)
```

```python
df
```

```python
# Group by Method is a method that is used for Aggrigations.

df.groupby('Furnishing')['Price'].sum()
```

```python
# There are some of the methods that we can use with Gropuby
# Sum, Count, value_counts, mean, median, mode.

df.groupby('Furnishing')['Price'].mean()
```

```python
# We can use more than one column by which we want to aggrigate another column.

df.groupby(['District','Location','Furnishing'])['Price'].mean()
```

```python
# Data Manipulation with Pandas 


# Data Sorting (Ascending and Descending Order).

# Sort Data According to Price.

df.sort_values(by='Price') # by Default Sort-values will work on ascending order.
```

```python
# Sort Values in Descending Order

df.sort_values(by='Price',ascending=False)
```

```python
# How to Create a New Column in Current Dataset.
# Thsi method will add a new column in the end.

df['discount_price'] =  df.Price - df.Price * 10/100
```

```python
# Pandas Concatination and Pandas merge methods for Data Wrangling.


admin = pd.DataFrame({'Name':['Shipra','Rupal','Rishi','Kaif','Kunal','Anil'],
                     'Age':[22,23,24,25,26,27],
                     'City':['Delhi','Delhi','Delhi','Delhi','Delhi','Delhi']})
```

```python
admin
```

```python
academics = pd.DataFrame({'Marks':[44,65,75,39,77,44],
                     'Grade':['c','b','b','d','a','c']})
```

```python
academics
```

```python
# Concatination

# Concatination will work on any data. (If there is not a matched column cocat still work on that)

# Concat will work on axis 0 by default (Adding Rows)

pd.concat([admin,academics])
```

```python

# Axis 1 is used to concat the data according to columns (Vertically)

pd.concat([admin,academics],axis=1)
```

```python
admin = pd.DataFrame({'Name':['Shipra','Rupal','Rishi','Kaif','Kunal','Anil'],
                     'Age':[22,23,24,25,26,27],
                     'City':['Delhi','Delhi','Delhi','Delhi','Delhi','Delhi']})

academics = pd.DataFrame({'Marks':[44,65,75,39,77],
                     'Grade':['c','b','b','d','a']})
```

```python
pd.concat([admin,academics],axis=1)
```

```python
admin = pd.DataFrame({'Roll_no':[1,2,3,4,5,6],
                      'Name':['Shipra','Rupal','Rishi','Kaif','Kunal','Anil'],
                     'Age':[22,23,24,25,26,27],
                     'City':['Delhi','Delhi','Delhi','Delhi','Delhi','Delhi']})

academics = pd.DataFrame({'Roll_no':[1,2,3,4,7,8],
                          'Marks':[44,65,75,39,77,33],
                     'Grade':['c','b','b','d','a','d']})
```
